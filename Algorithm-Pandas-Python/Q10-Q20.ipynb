### Container With Most Water
* **Medium**

- You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).
- Find two lines that together with the x-axis form a container, such that the container contains the most water.
- Return the maximum amount of water a container can store.
- Notice that you may not slant the container.

### Example 1:
- Input: height = [1,8,6,2,5,4,8,3,7]
- Output: 49
- Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of water (blue section) the container can contain is 49.
### Example 2:
- Input: height = [1,1]
- Output: 1

### Constraints:

- n == height.length
- 2 <= n <= 105
- 0 <= height[i] <= 104

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        
        # iki pointer ile başlayarak sağdan ve soldan duruma göre daralarak en büyük alanı bulucağız
        left = 0
        right = len(height) - 1
        max_water = 0
        
        # döngü sol sağdan küçük olduğu sürece devam edicektir
        while left < right:
            # a. Genişlik (Direkt fark) kısacası bizim iki düzlem olarak düşünürsek alt kısmıdır
            w = right - left
            
            # b. Yükseklik (Kısa olan belirler)  çünkü su taşar ve dökülür 
            # Python'un min fonksiyonu işi kolaylaştırır bu pointerlarımızın arraydeki karşığına bakılır ve min olan alınır
            h = min(height[left], height[right])
            
            # c. Alan Hesabı yapılır ve max olarak karşılaştırmayla en büyük alan güncellenir
            current_area = w * h
            max_water = max(max_water, current_area)
            
            # d. HAREKET ZAMANI bu kısım önemlidir eğer iki sütündan en kısası hareket etmelidir çünkü içerisinde (iki indeks arasında) daha uzun bir çubuğa denk gelebilriz 
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
                
        return max_water

a = Solution()
print(a.maxArea([1,8,6,2,5,4,8,3,7])) # Çıktı: 49 olmalı
```

### Integer to Roman
* **Medium**

- Seven different symbols represent Roman numerals: **I, V, X, L, C, D, M**
- Roman numerals are formed by appending values from highest to lowest decimal place.
- Conversion rules:

  * If the number does **not** start with 4 or 9, subtract the largest possible Roman value and append its symbol.
  * If the number **does** start with 4 or 9, use the subtractive form (e.g., 4 → IV, 9 → IX).
  * Only powers of 10 (**I, X, C, M**) can appear up to three times consecutively.
  * The allowed subtractive forms are: **IV (4), IX (9), XL (40), XC (90), CD (400), CM (900)**.
- Given an integer, convert it to a Roman numeral.

### Example 1:

* Input: `num = 3749`
* Output: `"MMMDCCXLIX"`
* Explanation:

  * 3000 → `MMM`
  * 700 → `DCC`
  * 40 → `XL`
  * 9 → `IX`

### Example 2:

* Input: `num = 58`
* Output: `"LVIII"`
* Explanation:

  * 50 → `L`
  * 8 → `VIII`

### Example 3:

* Input: `num = 1994`
* Output: `"MCMXCIV"`
* Explanation:

  * 1000 → `M`
  * 900 → `CM`
  * 90 → `XC`
  * 4 → `IV`

### Constraints:

* `1 <= num <= 3999`

```python
class Solution:
    def intToRoman(self, num: int) -> str:
        # 1. Hileli Haritamız (Büyükten küçüğe sıralı olması ŞART) çünkü roman sayısında en büyük olanı alıp çıkartıp kalan sayıdan devam edicek formattadır
        # Özel durumları (900, 400 vs.) normal bir sayı gibi ekledik. bu bizim hilemizdir ayrıca buraya bir algoritma kurmadık
        val_map = [
            (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"),
            (100, "C"), (90, "XC"), (50, "L"), (40, "XL"),
            (10, "X"), (9, "IX"), (5, "V"), (4, "IV"), (1, "I")
        ]
        
        # çıkarılan en büyük değerin ekleneceği en son dönücek olan sonucumuz
        res = ""
        
        # 2. Döngü: Haritadaki her değeri sırayla deneyecek önce 1000 değeri var  mı varsa çıkar ve sembolü res'e ekle sonra num değerinden bu 1000 çıkart eğer kalanda 1000 varsa devam et ve kalmayana kadar devamdan sonra diğer değerlere ge ve aynı mantığı uygula 
        for value, symbol in val_map:
            # 3. bu döngüde kontrol edip değeri çıkarıp res'e ekleme ve değeri numdan çıkarma yeridir
            while num>=value:
                res+=symbol
                num-=value
        
        return res
a = Solution()
print("MCMXCIV equal? :",a.intToRoman(1994)) # MCMXCIV
print("LVIII equal? :",a.intToRoman(58))   # LVIII
print("MMMDCCXLIX equal? :",a.intToRoman(3749)) # MMMDCCXLIX
```

### Roman to Integer

* **Easy**

- Roman numerals are represented by seven symbols: **I, V, X, L, C, D, M**

  | Symbol | Value |
  | ------ | ----- |
  | I      | 1     |
  | V      | 5     |
  | X      | 10    |
  | L      | 50    |
  | C      | 100   |
  | D      | 500   |
  | M      | 1000  |

- Rules for interpretation:

  * Roman numerals are generally written from **largest to smallest** left to right.
  * However, in **six subtractive cases**, a smaller numeral before a larger one indicates subtraction:

    * **IV (4), IX (9)**
    * **XL (40), XC (90)**
    * **CD (400), CM (900)**
  * Example:

    * *"IV"* → 4 because **I** before **V** means 5 − 1
    * *"IX"* → 9 because **I** before **X** means 10 − 1

- Given a Roman numeral string, convert it to an integer.

---

### Example 1:

* Input: `s = "III"`
* Output: `3`
* Explanation: `III = 3`

### Example 2:

* Input: `s = "LVIII"`
* Output: `58`
* Explanation:

  * L = 50
  * V = 5
  * III = 3

### Example 3:

* Input: `s = "MCMXCIV"`
* Output: `1994`
* Explanation:

  * M = 1000
  * CM = 900
  * XC = 90
  * IV = 4

---

### Constraints:

* `1 <= s.length <= 15`
* `s` contains only characters from `{'I','V','X','L','C','D','M'}`
* It is guaranteed that `s` is a valid Roman numeral in the range `[1, 3999]`


```python
class Solution:
    # dün ki problemi tersine mühendislik ile cozume ulastirdik roma rakamının içerisinde val_map ile başlıyorsa onun değeri kadar dönücek sonuca ekle ve bu symbolü roma rakımından çıkar mnatığı ile çözüldü
    def romanToInt(self, s: str) -> int:
        val_map = [
            (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"),
            (100, "C"), (90, "XC"), (50, "L"), (40, "XL"),
            (10, "X"), (9, "IX"), (5, "V"), (4, "IV"), (1, "I")
        ]
        
        res = 0
        
        for value, symbol in val_map:
            while s.startswith(symbol):
                res+=value
                s = s[len(symbol):]
        
        return res
a = Solution()
a.romanToInt("MCMXCIV") # 1994
```

### Longest Common Prefix

* **Easy**

- Given an array of strings, find the **longest shared starting substring** (prefix) among them.

- If the strings do **not** share any common prefix, return an empty string `""`.

---

### How It Works

To find the longest common prefix:

1. Compare characters of the strings **position by position**.
2. Stop when:

   * Characters differ, or
   * You reach the end of any string.
3. Everything matched up to that point is the longest common prefix.

Alternatively, sort the list and compare only the **first** and **last** strings — this works because sorting groups similar prefixes together.

---

### Example 1:

* Input: `strs = ["flower","flow","flight"]`
* Output: `"fl"`
* Explanation:
  All strings start with `"fl"` before diverging.

### Example 2:

* Input: `strs = ["dog","racecar","car"]`
* Output: `""`
* Explanation:
  There is **no** shared prefix among all strings.

---

### Constraints:

* `1 <= strs.length <= 200`
* `0 <= strs[i].length <= 200`
* Each string contains only **lowercase English letters** (if non-empty)


```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        res=""  # sonucu doneceğimiz string
        strs.sort() # öncelikle basit işlemle bütün stringleri alfabetik olarak sıralıyoruz amacımız ilk ve son nesneleri arasında en uzun prefixi bulursak aralarındakilerde aynı prefixe sahiptir

        s1=strs[0] # ilk nesne
        s2=strs[len(strs)-1] # son nesne
        
        for i in range(min(len(s1),len(s2))): # bu iki nesneden en kısa  olan hangisiyse onun karakter sayısıı kadar döngü devam etsin
            if s1[i]==s2[i]: # eğer bu indexteki karakterler birbrine eşitse sonuca ekle değilse res return et
                res+=s1[i]
            else:
                return res
            
        return res
a = Solution()
a.longestCommonPrefix(["flower","flow","floight"])
```

### 3Sum

* **Medium**

- Given an integer array `nums`, find **all unique triplets** `(i, j, k)` such that:

  * `i != j`, `i != k`, and `j != k`
  * `nums[i] + nums[j] + nums[k] == 0`
- The result **must not** contain duplicate triplets.

---

### How It Works

To find all unique triplets that sum to zero:

1. **Sort** the array to make it easier to avoid duplicates.
2. Loop through the array using an index `i`:

   * For each `nums[i]`, use the **two-pointer technique** on the subarray to the right.
3. Move the left and right pointers inward based on whether the sum is:

   * Greater than 0 → decrease the right pointer.
   * Less than 0 → increase the left pointer.
   * Exactly 0 → record the triplet.
4. **Skip duplicates** for both the main index and the pointers to ensure uniqueness.

---

### Example 1:

* Input: `nums = [-1,0,1,2,-1,-4]`
* Output: `[[-1,-1,2], [-1,0,1]]`
* Explanation:
  Valid triplets that sum to zero are:

  * `(-1) + (-1) + 2 = 0`
  * `(-1) + 0 + 1 = 0`

---

### Example 2:

* Input: `nums = [0,1,1]`
* Output: `[]`
* Explanation:
  No combination of three numbers sums to zero.

---

### Example 3:

* Input: `nums = [0,0,0]`
* Output: `[[0,0,0]]`
* Explanation:
  The only possible triplet sums to zero.

---

### Constraints:

* `3 <= nums.length <= 3000`
* `-10⁵ <= nums[i] <= 10⁵`


```python
from typing import List


class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = []
        # 1. Sıralama  yapmak durumundayız çünkü hangi yöne doğru büyüdüğünü anlamazsak complexity arttırarak çözmek durumunda kalırız
        nums.sort()
        
        n = len(nums) # listenin uzunluğu
        
        for i in range(n): # bizim bir sayıyı sabitleyip diğer iki sayıyının toplamlarının bu sayının negatifine eşit olmasını isteyeceğiz
            # 2. Duplicate Kontrolü yapmaız gerkiyor eğer bir sonraki değer bizim şuanda tutuğumuz der ise bunu atlayabilriiz demektir çübkü biz unique değerler arıyoruz
            # Eğer sayı bir öncekiyle aynıysa, tekrar hesaplamaya gerek yok.
            if i > 0 and nums[i] == nums[i-1]:
                continue
            
            # 3. Two Pointer Kurulumu birer left ve right flagleri ile sağ sola kaydırarak kombinasyonları deniyoruz
            target = -nums[i] # Aradığımız toplam hedef
            left = i + 1      # Sabit sayının hemen sağı
            right = n - 1     # En son eleman
            
            while left < right: # sağdaki indeks soldakinden büyük olduğu sürece devam etsin
                current_sum = nums[left] + nums[right] # hedefimizle karlışıalştıracağımız elimizdeki anlık kombinasyon sonucundaki sayı
                
                if current_sum == target: # bu eşitlik sağlanırsa kombinasyon bulundu
                    res.append([nums[i], nums[left], nums[right]]) # ve sonuca ekledik
                    
                    # Sonsuz döngüye girmemek için left ve right oynamalı. ki diğer kombinasyonlrı bulabilelim 
                    # Ayrıca [0, -1, -1, 2, 2] gibi durumlarda, aynı sayıları (duplicate) atlayarak ilerlemeliyiz çünkü aynı sayından birden fazla olabilir
                    while left < right and nums[left] == nums[left+1] :
                        left+=1
                    
                    while left < right and nums[right] == nums[right-1] :
                        right-=1
                    
                    # Pointerları kaydırıyoruz
                    left += 1
                    right -= 1
                    
                # bu kısmı eğer şuan elimizdeki sayı targettan küçükse elimzideki kombinasyonun küçük elemanını büyütmeye çalışırız bu da en küçük indeksi tutan elemanımız olur aynı şekilde tam tersi mantıkla da target büyükse diye düşünmeliyiz
                elif current_sum < target:
                    # Toplam çok küçük, büyütmek lazım
                    left += 1
                else:
                    # Toplam çok büyük, küçültmek lazım
                    right -= 1
                    
        return res
    
a=Solution()
print(a.threeSum([-1,0,1,2,-1,-4]))
print(a.threeSum([0,0,0]))
print(a.threeSum([0,1,1]))
```
