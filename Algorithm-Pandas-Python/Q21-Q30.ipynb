### Merge Two Sorted Lists

* **Easy**

- You are given the heads of two sorted linked lists `list1` and `list2`.
- Merge the two lists into one sorted list.
- The list should be made by splicing together the nodes of the first two lists.
- Return the head of the merged linked list.

### Example 1:

* **Input:** `list1 = [1,2,4], list2 = [1,3,4]`
* **Output:** `[1,1,2,3,4,4]`

### Example 2:

* **Input:** `list1 = [], list2 = []`
* **Output:** `[]`

### Example 3:

* **Input:** `list1 = [], list2 = [0]`
* **Output:** `[0]`

### Constraints:

* The number of nodes in both lists is in the range `[0, 50]`.
* `-100 <= Node.val <= 100`
* Both `list1` and `list2` are sorted in non-decreasing order.


```python
from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # geri donus yapacağımız listnode icin onune hayali bir node koyuyoruz head yaratıyoruz boş bir tane sornasında kuyruk sistemiyle kuyruğa küçükler eklenmeye devam ediyor
        dummy = ListNode()
        tail = dummy

        # reecursive yapı misali list1 ve list2 node var olduğu sürece devam etsin logic
        while list1 and list2:
            # eğer list1in ki list2ninkinden küçükse kuyruğa list1 i ekle ve list1 nodenun headi artık bir sonraki olsun 
            # 1->2->3-> // 2->3 olmasını sağlayan ve dumy için kuyruğa 1 ekleyen kısım aynı mantık else de list 2 için var
            if list1.val < list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            
            # tail ilerletmeliyiz ki kopardığımız nodeları sonrasına bağlyabilelim
            tail=tail.next
            
        # eğer list1 veya 2 varsa sonraki node diretk ondan oluşur yani
        if list1:
            tail.next = list1
        elif list2:
            tail.next = list2
            
        # boş atadığımız nodun sonrası bizim head kısmımız            
        return dummy.next

a=Solution()
solve=a.mergeTwoLists(ListNode(1,ListNode(2,ListNode(3,ListNode(4,ListNode(5))))),ListNode(3,ListNode(4,ListNode(5,ListNode(5,ListNode(5))))))

current = solve
while current:
    print(current.val, end=" -> ")
    current = current.next
```
