### Merge Two Sorted Lists

* **Easy**

- You are given the heads of two sorted linked lists `list1` and `list2`.
- Merge the two lists into one sorted list.
- The list should be made by splicing together the nodes of the first two lists.
- Return the head of the merged linked list.

### Example 1:

* **Input:** `list1 = [1,2,4], list2 = [1,3,4]`
* **Output:** `[1,1,2,3,4,4]`

### Example 2:

* **Input:** `list1 = [], list2 = []`
* **Output:** `[]`

### Example 3:

* **Input:** `list1 = [], list2 = [0]`
* **Output:** `[0]`

### Constraints:

* The number of nodes in both lists is in the range `[0, 50]`.
* `-100 <= Node.val <= 100`
* Both `list1` and `list2` are sorted in non-decreasing order.


````python
from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # geri donus yapacağımız listnode icin onune hayali bir node koyuyoruz head yaratıyoruz boş bir tane sornasında kuyruk sistemiyle kuyruğa küçükler eklenmeye devam ediyor
        dummy = ListNode()
        tail = dummy

        # reecursive yapı misali list1 ve list2 node var olduğu sürece devam etsin logic
        while list1 and list2:
            # eğer list1in ki list2ninkinden küçükse kuyruğa list1 i ekle ve list1 nodenun headi artık bir sonraki olsun 
            # 1->2->3-> // 2->3 olmasını sağlayan ve dumy için kuyruğa 1 ekleyen kısım aynı mantık else de list 2 için var
            if list1.val < list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            
            # tail ilerletmeliyiz ki kopardığımız nodeları sonrasına bağlyabilelim
            tail=tail.next
            
        # eğer list1 veya 2 varsa sonraki node diretk ondan oluşur yani
        if list1:
            tail.next = list1
        elif list2:
            tail.next = list2
            
        # boş atadığımız nodun sonrası bizim head kısmımız            
        return dummy.next

a=Solution()
solve=a.mergeTwoLists(ListNode(1,ListNode(2,ListNode(3,ListNode(4,ListNode(5))))),ListNode(3,ListNode(4,ListNode(5,ListNode(5,ListNode(5))))))

current = solve
while current:
    print(current.val, end=" -> ")
    current = current.next
````

### Generate Parentheses

* **Medium**

- You are given an integer `n` representing the number of pairs of parentheses.
- Generate **all** combinations of **well-formed** parentheses consisting of exactly `n` pairs.
- Return the combinations in any order.

### Example 1:

* **Input:** `n = 3`
* **Output:** `["((()))","(()())","(())()","()(())","()()()"]`

### Example 2:

* **Input:** `n = 1`
* **Output:** `["()"]`

### Constraints:

* `1 <= n <= 8`

````python
# mantığımız kombinasyondur lakin bu şartlı bir kombinasyondur sistemimiz şu şekilde olmalıdır:

# eğer Durma Noktası (Base Case): Parantez sayısı hedefe ulaştı mı? (2 * n uzunluğunda mı?) uzunluğundaysa bunu dönüceğimiz listeye ekliyoruz

# Kural 1 (Açma): Elimizde hala ( hakkı var mı? (acik < n) varsa koyabiliriz istediğimiz zaman

# Kural 2 (Kapama): İçeride kapatılmayı bekleyen bir ( var mı? (kapali < acik)   # varsa o zaman kapama koyabiliriz yoksa koyamayız

from typing import List

class Solution:
    def generateParenthesis(self, n: int) :
        liste=[]
        
        def backtrack(open_c,close_c,current_str):
            if len(current_str)==n*2:
                liste.append(current_str)
                return

            if open_c<n:
                backtrack(open_c+1,close_c,current_str+"(")
            
            if close_c < open_c :
                backtrack(open_c,close_c + 1,current_str+")")
        backtrack(0,0,"")

        return liste , len(liste)
a=Solution()
print(a.generateParenthesis(4))
````

### Merge k Sorted Lists

* **Hard**

* You are given an array of `k` linked-lists `lists`, each linked-list is sorted in ascending order.
* Merge all the linked-lists into one sorted linked-list and return it.

### Example 1:

* **Input:** `lists = [[1,4,5],[1,3,4],[2,6]]`
* **Output:** `[1,1,2,3,4,4,5,6]`
* **Explanation:** The linked-lists are:
```
[
  1 -> 4 -> 5,
  1 -> 3 -> 4,
  2 -> 6
]

```


merging them into one sorted linked list:
`1 -> 1 -> 2 -> 3 -> 4 -> 4 -> 5 -> 6`

### Example 2:

* **Input:** `lists = []`
* **Output:** `[]`

### Example 3:

* **Input:** `lists = [[]]`
* **Output:** `[]`

### Constraints:

* `k == lists.length`
* `0 <= k <= 10^4`
* `0 <= lists[i].length <= 500`
* `-10^4 <= lists[i][j] <= 10^4`
* `lists[i]` is sorted in **ascending order**.
* The sum of `lists[i].length` will not exceed `10^4`.

````python
from typing import Optional , List


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def create_linked_list_from_array(arr):
        if not arr: return None
        head = ListNode(arr[0])
        curr = head
        for val in arr[1:]:
            curr.next = ListNode(val)
            curr = curr.next
        return head

class Solution:
    # onceki günlerdeki iki sıralı linked listi birleştiren fonksiyon işimize yarayacağı için kullanıyoruz
    def mergeTwoLists(self,list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        tail = dummy

        while list1 and list2:
            if list1.val < list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            
            tail=tail.next
            
        if list1:
            tail.next = list1
        elif list2:
            tail.next = list2
            
        return dummy.next
    

    # ANA PROBLEM BURADADIR
    # sistem şu şekilde işlemek durumundadır listenin ilk iki listnode alıp bunları birleştiren fonksiyona vermeyi hedefler ana hattıyla
    # sonrasında elde edileni geçici bir listeye atarız sonrasında bu listeyi parametre olarak verilen listeye eşitleriz
    # eğer hiç elemanı yoksa none döneriz lakin varsada sistem ikişer ikişer ilerleyerek sıradaki birleştireceklerini bir recursive mantığa benzer şekilde işler bütün listeyi
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        # Kenar Durum: Liste boşsa
        if not lists or len(lists) == 0:
            return None
        
        # Tek bir liste kalana kadar döngüye devam ediyoruz
        while len(lists) > 1:
            merged_lists = []
            
            # Listeleri 2'şer 2'şer gezmemiz lazım
            for i in range(0, len(lists), 2):
                l1 = lists[i]
                # Dikkat: l2 var mı, Listenin sayısı tek olabilir bu durumda none deriz ve brileşme sonrasında aslında l1 donmus oluruz
                l2 = lists[i+1] if (i + 1) < len(lists) else None
                
                # Şimdi bu ikisini mergeTwoLists ile birleştirip merged_lists dizisine ekliyoruz
                temp=self.mergeTwoLists(l1,l2)
                merged_lists.append(temp)

            lists = merged_lists # dongu kırılmasın diye bulduğumuz son birleşmişlerden oluşan linked listi dondureceğimiz değişkene atıyoruz 
            # ornek merged_list={[l1 ve l2 birleşimi] , [l3 ve l4 birleşimi]}  bunların tekrar birleştirilmesi lazım ki sistem tek birleştirilmiş olarak dönsün
            
        return lists[0]



# 1. Test verilerini Linked List'e çevirelim 
data = [[1,4,5], [1,3,4,10], [2,6,8]]
linked_lists_input = [ListNode.create_linked_list_from_array(lst) for lst in data]

# 2. Çözümü Çalıştırıyoruz
a = Solution()
solve = a.mergeKLists(linked_lists_input)

# 3. Sonucu Yazdırıyoruz
print("Sonuç:")
current = solve
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
````

### Swap Nodes in Pairs

* **Medium**

* Given a linked list, swap every two adjacent nodes and return its head.
* You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed).

### Example 1:

* **Input:** `head = [1,2,3,4]`
* **Output:** `[2,1,4,3]`

### Example 2:

* **Input:** `head = []`
* **Output:** `[]`

### Example 3:

* **Input:** `head = [1]`
* **Output:** `[1]`

### Example 4:

* **Input:** `head = [1,2,3]`
* **Output:** `[2,1,3]`

### Constraints:

* The number of nodes in the list is in the range `[0, 100]`.
* `0 <= Node.val <= 100`

````python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def create_linked_list_from_array(arr):
        if not arr: return None
        head = ListNode(arr[0])
        curr = head
        for val in arr[1:]:
            curr.next = ListNode(val)
            curr = curr.next
        return head

class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 1-2-3 : 2-1-3 olsun istiyoruz bu durumda iki flag(pointer mantığıyla ilerleyip bu iksii arasında 2 node 
        # varsa devam ediceğiz değiştirmeye diye sistem kurucağız)
        # bizim geçici nodeuumuz headi tutucak en son dönerken linkedlistin başını işaret etmek için 
        dummy = ListNode(0, head)
        
        # bu kısımda da bizim pointerlerımız olacak olan harekt ettirirken kontrol noktalarımız olucak "prev" dir
        prev = dummy
        
        # yai iki node varsa sırasıyla değiştirmeye devam ediceğiz
        while prev.next and prev.next.next:
            
            first = prev.next       # Şu anki 1. düğüm (Örn: 1)
            second = first.next     # Şu anki 2. düğüm (Örn: 2)
            
            # ilk değişim 1.yi gösteren kısım artık 1 yerine 2yi göstermek için 2yi işaret etti sistem şu şekilde oldu
            # prev->2 1->2->3
            prev.next = second 
            
            # 1in artık 2 yi göstermesi gerekemz yer değiştirdiği için 2nin sağındakini yani 3 göstermeli
            # prev->2  1->3  2->3 gözükmesi lazımdır ben prev.next.next daha makul geldi ama okunaklılık için yorumdaki kısım geçebilir
            first.next = prev.next.next # second.next
            
            # son gösterim sonrsında farkettiyseniz artık 2in gösterdiği kısım 3 bu bizim linkedlisti bozar 2 artık 1 göstermeli
            # prev->2->1->3 şeklinde tamamlandı sistemimiz 
            second.next = first

            # dongumuz kırılmaması için previn iki adet node atlaması lazım oda prev artık 3ü gösteren node üzerinde olmalı ki döngü devam ederse
            # 3 ve 3ün gösterdiği node değişim yapabilsin diye
            prev = first 
            
        return dummy.next
    
data = [1,2,3,4,9,8,5,6]  # 2-1-4-3-8-9-6-5 şeklinde olmalıdır
linked_lists_input = ListNode.create_linked_list_from_array(data)

a = Solution()
solve = a.swapPairs(linked_lists_input)
# 3. Sonucu Yazdırıyoruz
print("Sonuç:")
current = solve
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
````

### Reverse Nodes in k-Group

* **Hard**
* Given the `head` of a linked list, reverse the nodes of the list `k` at a time, and return the modified list.
* `k` is a positive integer and is less than or equal to the length of the linked list. If the number of nodes is not a multiple of `k` then left-out nodes, in the end, should remain as it is.
* You may not alter the values in the list's nodes, only nodes themselves may be changed.

### Example 1:

* **Input:** `head = [1,2,3,4,5], k = 2`
* **Output:** `[2,1,4,3,5]`

### Example 2:

* **Input:** `head = [1,2,3,4,5], k = 3`
* **Output:** `[3,2,1,4,5]`

### Constraints:

* The number of nodes in the list is `n`.
* `1 <= k <= n <= 5000`
* `0 <= Node.val <= 1000`

````python
from typing import Optional

# klasik listnode tanımlamalarımız
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def create_linked_list_from_array(arr):
        if not arr: return None
        head = ListNode(arr[0])
        curr = head
        for val in arr[1:]:
            curr.next = ListNode(val)
            curr = curr.next
        return head

    class Solution:
        def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
            """ Bu fonksiyon da öncelikle verilen k kadar bir aralığı keşfetmemiz gerekiyor bir listnode içeriside baştan başlayarak k mesafeli diğer nodu bulmalıyız öncelikle ardından o o knın nextini tutmalıyız ki dğeişim olduğunda terslediğimiz sublisti buraya bağlayalım bu durumda tersini alarak sonraki node bağlayıp aynı işlemleri sonraki sublist için yapıyoruz"""
            dummy = ListNode(0, head)  # geçici larak listenin başını tutuyoruz ki return ederekn dummy.next ile listenin head verebilelim
            group_prev = dummy # grubun öncesini tutmamız lazım ki k'ncı node bulurken hesaplayabilelim

            while True:
                kth = self.get_kth(group_prev, k) # k'nıncı node bulundu
                if not kth: # k'nıncı node yoksa sdöngü kırılır ve var olan dönülür
                    break

                group_next = kth.next # sub listi terse çevirdiğimizde sıradakine bğalmamaız lazım bu yüzden şuan da var olan sublistin knıncısının nextini tutuyoruz

                group_start = group_prev.next # listeyi döndürmeye başlayacağımız yer yani "head" 

                new_group_head, next_after_group = self.reverseList(group_start, group_next) # terse döndürme yapıldı ve k(new grou head) - head(new after group) kısmı oldu

                group_prev.next = new_group_head  # grubun başına döndürdüğümüz kısmı getirdik

                group_start.next = next_after_group  # grubun yeni başlangıcı artık bu tersine döndürdüğümüz kısımdan sonrasıdır

                group_prev = group_start # grup previ strta atıyoruz ki döngü devam etsin sıradaki subliste geçsin

            return dummy.next

        def get_kth(self, curr: Optional[ListNode], k: int) -> Optional[ListNode]:
            while curr and k > 0:  # şuan var olan ve k 0dan büyükse devam edelim
                curr = curr.next  # curn bir sonrakinne atıyoruz k.ncı bulunana kadar devam  
                k -= 1 # k yı azaltıyoruz ki sayabilelim
            return curr

        def reverseList(self, start: ListNode, group_next: Optional[ListNode]):
            """
            [start, group_next) aralığını tersine çevirir (group_next dahil değil).
            Döndürür:
            - new_head (eski k ): ters çevrilmiş parçanın yeni başı
            - curr: group_next (yani ters çevrilen parçadan sonraki node)
            """
            prev = group_next 
            curr = start

            while curr != group_next:
                tmp = curr.next
                curr.next = prev
                prev = curr
                curr = tmp

            return prev, curr



data = [1,2,3,4,5,6,7,8,9]  # 2 - 1 - 4 - 3 - 6 - 5 - 8 - 7 - 9 - None olmalıdır
linked_lists_input = ListNode.create_linked_list_from_array(data)

a = Solution()
solve = a.reverseKGroup(linked_lists_input,2)
# 3. Sonucu Yazdırıyoruz
print("Sonuç:")
current = solve
while current:
    print(current.val, end=" - ")
    current = current.next
print("None")
````
